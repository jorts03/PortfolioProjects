```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)


# Read in data
df = pd.read_csv(r'C:\Users\jcede\Downloads\movies.csv')
```

```python
df.head()
```

```python
# Check for missing data
df.isnull().sum()
```

```python
# Drop rows with missing data for simplicity
df = df.dropna(how='any', axis=0)
df.isnull().sum()
```

```python
# Data cleaning

# clean up budget and gross amounts
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['year_released'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
# Sort data by largest gross
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df.head()
```

```python
#Budget vs Gross
plt.scatter(x = df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.ylabel('Gross Earnings')
plt.xlabel('Budget for Film')
plt.show()
```

```python
# PLot vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Correlation between budget, gross, runtime, score, votes
df[['budget','gross','runtime','score','votes']].corr(method = 'pearson')
```

```python
correlation_matrix = df[['budget','gross','runtime','score','votes']].corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

```python
# Change text to numeric
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
# Correlation for every field
correlation_matrix_numerized = df_numerized.corr(method = 'pearson')
sns.heatmap(correlation_matrix_numerized, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
# Highest Correlations

correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
